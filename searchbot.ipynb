```python
import requests
from bs4 import BeautifulSoup
import nltk

# Ensure you have the necessary NLTK data
nltk.download('punkt')

# Google Custom Search API credentials
API_KEY = 'AIzaSyC5_xiCQtuVXiZ9YwElH6iacxNSMO9OrFA'
SEARCH_ENGINE_ID = '05ed93df7e5674baa'
QUERY=input("Input your query")

def fetch_search_results(query, api_key, search_engine_id, num_results=10):
    try:
        url = f'https://www.googleapis.com/customsearch/v1?q={query}&key={api_key}&cx={search_engine_id}&num={num_results}'
        response = requests.get(url)
        response.raise_for_status()
        return response.json()
    except Exception as e:
        return "hello"

def extract_main_content(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        paragraphs = soup.find_all('p')
        content = ' '.join([para.get_text() for para in paragraphs])
        return content
    except Exception as e:
        return "hello"

def combine_words(contents):
    try:
        tokenized_contents = [nltk.word_tokenize(content) for content in contents]
        combined_words = []

        max_length = max(len(tokens) for tokens in tokenized_contents)
    
        for i in range(max_length):
            for tokens in tokenized_contents:
                if i < len(tokens):
                    combined_words.append(tokens[i])
    
        return ' '.join(combined_words)

    except Exception as e:
        return "hello"

search_results = fetch_search_results(QUERY, API_KEY, SEARCH_ENGINE_ID)
contents = []

for item in search_results.get('items', []):
        link = item.get('link')
        try:
            content = extract_main_content(link)
            contents.append(content)
        except Exception as e:
            print(f"Error fetching content from {link}: {e}")
        
combined_string = combine_words(contents)
```

```python
import requests
def get_search_results(query, num_pages=3):
    results = []
    for page in range(1, num_pages+1):
        url = f"https://api.duckduckgo.com/?q={query}&format=json&skip={((page-1)*10)}"
        response = requests.get(url)
        data = response.json()
        if 'RelatedTopics' in data:
            results.extend(topic['Text'] for topic in data['RelatedTopics'] if 'Text' in topic)
        else:
            break
    return results

def combine_words_from_search_results(query, num_pages=3):
    search_results = get_search_results(query, num_pages)
    combined_string = ' '.join(search_results)
    return combined_string
num_pages = 3
result = combine_words_from_search_results(QUERY, num_pages)
result = result +combined_string
```
